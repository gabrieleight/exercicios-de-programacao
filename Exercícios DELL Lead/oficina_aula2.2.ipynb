## Quest√£o 01

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
games_uri = 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro04/oficina/arquivos/dadosoficina04.txt'
games = pd.read_csv(games_uri)
```

```python
games.head()
```

```python
generos = games.groupby(by='Genre').sum()
generos
```

## Quest√£o 02

```python
plt.figure(figsize=(12, 12))
plt.style.use('seaborn-whitegrid')
plt.title('Propor√ß√£o de vendas globais por g√™nero')
plt.pie(x=generos['Global_Sales'], labels=generos.index, autopct='%1.1f%%', startangle=90, pctdistance=0.85)
my_circle = plt.Circle((0, 0), 0.7, color='white')
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

## Quest√£o 03

```python
games_2015 = games['Year'] >= 2015
games_filtered = games[games_2015]
games_filtered
```

```python
plataforma = games_filtered.groupby(by='Platform').count()
plataforma
```

```python
plataforma = plataforma.sort_values(by='Year', ascending=False)
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(20,10))
plt.title('Propor√ß√£o de quantidade de jogos por plataforma, a partir de 2015\n')
plt.bar(plataforma.index, plataforma['Rank'])
plt.xlabel('Plataforma')
plt.ylabel('Quantidade\n')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Exemplo 1

```python
especie = ['Cachorro', 'Gato', 'Pássaro']
quantidade = [110, 90, 30]
plt.figure(figsize=(12, 12))
plt.style.use('seaborn-whitegrid')
plt.title('Proporção de espécies')
plt.pie(x=quantidade, labels=especie, autopct='%1.2f%%', colors='rgb')
plt.show()
```

## Exemplo 4

```python
games_uri = 'https://raw.githubusercontent.com/cabrau/curso-analise-visualizacao-dados-python/master/vgsales.csv'
games = pd.read_csv(games_uri)
games.head()
```

```python
generos = games.groupby(by='Genre').sum()
generos
```

```python
plt.figure(figsize=(12, 12))
plt.style.use('seaborn-whitegrid')
plt.title('Proporção de vendas globais por gênero')
plt.pie(x=generos['Global_Sales'], labels=generos.index, autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(12, 12))
plt.style.use('seaborn-whitegrid')
plt.title('Proporção de vendas globais por gênero')
plt.pie(x=generos['Global_Sales'], labels=generos.index, autopct='%1.1f%%', startangle=90, pctdistance=0.85)
my_circle = plt.Circle((0, 0), 0.7, color='white')
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
generos = generos.sort_values(by='Global_Sales', ascending=False)
porcentagens = generos['Global_Sales']*100.0 / generos['Global_Sales'].sum()
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(15,5))
plt.title('Proporção de vendas de jogos por gênero\n')
plt.bar(generos.index, porcentagens)
plt.ylabel('Porcentagem (em %)\n')
plt.show()
```

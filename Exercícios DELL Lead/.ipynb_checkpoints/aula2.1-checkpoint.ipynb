```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Esse comando mágico do jupyter permite que as imagens geradas pelo matplotlib sejam geradas e exibidas no próprio notebook
%matplotlib inline  

#Código para criar um gráfico em barras

categorias = ['A', 'B', 'C', 'D']
valores = [10, 20, 15, 5]
# a função bar define que será gerado um gráfico de barras
plt.bar(categorias, valores)
# a função 'xlabel' define o nome do eixo x (horizontal) no gráfico
plt.xlabel('Grupos')
# a função 'ylabel' define o nome do eixo y (vertical) no gráfico
plt.ylabel('Tamanho')
# a função title define o título do gráfico
plt.title('Primeiro gráfico de barras')
# a função show mostra o gráfico no script/notebook
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline  


categorias = ['A', 'B', 'C', 'D']
valores = [10, 20, 15, 5]

# A função figure permite alterar diversos parâmentros da figura, em especial o seu tamanho pelo parâmetro figsize por meio de uma tupla
# A tupla recebe um par de valores (comprimento, altura)
plt.figure(figsize=(9,3))

# a função bar recebe o argumento 'darkred' no parâmetro color, como controle da estética da visualização
plt.bar(categorias, valores, color='darkred')
plt.xlabel('Grupos')
plt.ylabel('Tamanho')
plt.title('Primeiro gráfico de barras')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

filmes_uri = 'https://raw.githubusercontent.com/emanueles/datavis-course/master/assets/files/observable/movies.json'
filmes = pd.read_json(filmes_uri)


#print(filmes.head())
#print()
#print(filmes.describe())
#print()

# A função groupby tabula um grupo de dados categóricos em razão dos dados quantitativos
contagem = filmes.groupby(by='Genre').count()
print(contagem)

plt.figure(figsize=(9,5))
plt.bar(contagem.index, contagem["Worldwide_Gross_M"])
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Contagem')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

filmes_uri = 'https://raw.githubusercontent.com/emanueles/datavis-course/master/assets/files/observable/movies.json'
filmes = pd.read_json(filmes_uri)

# A função set recebendo o argumento 'whitegrid' no parâmetro style, define como o gráfico vai ser construído (no caso, sua visualização/estilo)
sns.set(style="whitegrid")
# A função countplot define como vai ser esse gráfico, a partir de um DataFrame, em que o parâmetro x (coluna dos dados categóricos)
# O parâmetro data recebe o DataFrame para quantificar os dados categóricos.
plt.figure(figsize=(9,3))
plt.title("Quantidade de filmes por gênero")
a = sns.countplot(x="Genre", color='g', data=filmes)
plt.show(a)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

filmes_uri = 'https://raw.githubusercontent.com/emanueles/datavis-course/master/assets/files/observable/movies.json'
filmes = pd.read_json(filmes_uri)


sns.set(style="whitegrid")
plt.figure(figsize=(9,3))
plt.title("Quantidade de filmes por gênero")
b = sns.countplot(y="Genre", color='r', data=filmes)
plt.show(b)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

filmes_uri = 'https://raw.githubusercontent.com/emanueles/datavis-course/master/assets/files/observable/movies.json'
filmes = pd.read_json(filmes_uri)


sns.set(style="whitegrid")
plt.figure(figsize=(9,3))
plt.title("Bilheteria média por gênero")
sns.barplot(x="Genre", y="Worldwide_Gross_M", color = 'b', ci = 'sd', data=filmes)
plt.show()  
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

filmes_uri = 'https://raw.githubusercontent.com/emanueles/datavis-course/master/assets/files/observable/movies.json'
filmes = pd.read_json(filmes_uri)


sns.set(style="whitegrid")
plt.figure(figsize=(9,3))
plt.title("Bilheteria total por gênero")
sns.barplot(x="Genre", y="Worldwide_Gross_M", color = 'orange', estimator = sum, ci = None, data=filmes)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

filmes_uri = 'https://raw.githubusercontent.com/emanueles/datavis-course/master/assets/files/observable/movies.json'
filmes = pd.read_json(filmes_uri)


plt.figure(figsize=(15,8))
plt.title("Filmes por gênero em cada ano")
sns.countplot(x="Year", hue="Genre", data=filmes)
plt.show()
```

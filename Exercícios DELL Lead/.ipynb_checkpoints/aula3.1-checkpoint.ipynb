```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
houses_uri = 'https://leadfortaleza.com.br/ead2pcd/conteudo/tmp/myopenolat_1_103418176718354/aula/arquivos/dados.txt'
houses = pd.read_csv(houses_uri)
```

## Exemplo 1

```python
houses.head()
```

```python
houses['area'].describe()
```

```python
houses['area']
```

```python
plt.figure(figsize=(12,5))
plt.style.use('seaborn-whitegrid')
plt.title('Distribuição de áreas dos imóveis\n')
plt.hist(houses['area'], bins=25)
plt.show()
```

## Exemplo 2

```python
plt.figure(figsize=(12, 5))
plt.title('Distribuição de áreas dos imóveis\n')
sns.distplot(houses['area'])
plt.xlabel('Área (m²)')
plt.ylabel('Densidade\n')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.title('Distribuição de áreas dos imóveis\n')
sns.kdeplot(houses['area'], shade=True, legend=False)
plt.xlabel('Área (m²)')
plt.ylabel('Densidade\n')
plt.show()
```

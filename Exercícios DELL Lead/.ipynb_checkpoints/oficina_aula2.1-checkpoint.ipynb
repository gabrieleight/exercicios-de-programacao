## Questão 01

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
filmes_uri = 'https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro03/oficina/arquivos/dadosoficina03.txt'
filmes = pd.read_json(filmes_uri)
filmes.head()
```

## Questão 02

```python
plt.figure(figsize=(20, 10))
plt.title('Orçamento de filmes\n')
sns.set(style='whitegrid')
sns.barplot(x='Budget_M', y='Film', data=filmes, palette='pastel')
plt.xlabel('Orçamento (em milhões de dólares)')
plt.ylabel('Filme\n')
plt.show()
```

## Questão 03

```python
plt.figure(figsize=(20, 10))
plt.title('Bilheteria mundial de filmes\n')
sns.set(style='whitegrid')
sns.barplot(x='Worldwide_Gross_M', y='Film', data=filmes, palette='pastel')
plt.xlabel('Bilheteria (em milhões de dólares)')
plt.ylabel('Filme\n')
plt.show()
```

## Questão 04

```python
plt.figure(figsize=(15, 8))
plt.title('Quantidade de filmes por gênero\n')
sns.set(style='whitegrid')
sns.countplot(y='Genre', palette='pastel', data=filmes)
plt.xlabel('Quantidade')
plt.ylabel('Gênero\n')
plt.show()
```

## Questão 05

```python
plt.figure(figsize=(15, 8))
plt.title('Bilheteria média por gênero\n')
sns.barplot(x='Genre', y='Worldwide_Gross_M', palette='pastel', ci=None, data=filmes)
plt.xlabel('Gênero')
plt.ylabel('Bilheteria média (em milhões de dólares)\n')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.title('Média do orçamento por gênero em cada ano\n')
sns.barplot(x='Year', y='Budget_M', palette='pastel', ci=None, hue='Genre', data=filmes)
plt.xlabel('Ano')
plt.ylabel('Orçamento (em milhões de dólares)\n')
plt.show()
```

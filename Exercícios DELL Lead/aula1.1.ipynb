```python
import pandas as pd

# Função Series tabula uma lista numérica em índices

s = pd.Series([10,20,30,40,50])

print(s)

# Função values retorna uma array a uma variável a lista que foi tabulada antes

r = s.values

print(r)

print(r[2])
```

```python
import pandas as pd

# Exemplo 1:

ss = pd.Series([15,20,12,16,10,11,12,15,9,10,12,17], index = ['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez'])
print(ss)
## Seleção de um dado de acordo com o index
print()
print('mês selecionado: ',ss['jun'])
## Seleção dos seis primeiros meses dos dados
print()
print(ss[0:6])
## Função para média da lista 'ss' tabulada
print()
media = ss.mean()
print(media)
## Função para determinar o desvio padrão da lista 'ss' tabulada
print()
dp = ss.std()
print(dp)
## Função para determinar o valor máximo e mínimo da lista 'ss' tabulada
print()
print(ss.max())
print(ss.min())
## Para determinar todas as estatísticas de uma só vez, use a função .describe()
print()
print(ss.describe())
## Para determinar o somatório dos acidentes ao longo dos meses, use a função .sum()
print()
print(ss.sum())
## Para determinar a média dos últimos seis meses da lista 'ss' tabulada
print()
print((ss[6:]).mean())
```

```python
import pandas as pd

# Tabulação de um dicionãrio por meio da função pd.DataFrame()
dataFrame = pd.DataFrame({'Aluno' : ['Marina','Felipe','Cleyton','Isabel'], 'Créditos cursados': [20,64,32,24], 'Rendimento acadêmico' : [8.55,7.88,8.17,9.04], 
                        'Mês de nascimento' : ['Novembro','Setembro','Janeiro','Julho'], 'Curso': ['Computação','Estatística','Computação','Matemática']})
print(dataFrame)

# Acessando aos dados de um aluno pelo index do Data Frame, usando a função .iloc[]
print()
print(dataFrame.iloc[2]) 

# Lembre que quando selecionamos uma variável do Data Frame, ela retorna uma Series, sendo possível utilizar as finções da Series
print()
rend = dataFrame['Rendimento acadêmico'].values
print(rend)
print(rend.mean())
print(rend.std())

# É possivel usar a função .describe() na Data Frame toda, apenas nas variáveis quantitativas
print()
print(dataFrame.describe())

#                        
```

```python
import pandas as pd

# Leitura de dados via Data Frame a partir de dados existentes na internet (ou localmente)
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/alcohol-consumption/drinks.csv')
## É possível fazer a leitura do arquivo csv direto pelo PC, desde que o arquivo esteja na mesma pasta
## df=pd.read_csv('drinks.csv')
print(df)

# A função .head(number) permite a exibição de uma parte da tabela, uma vez que ela é grande
print()
print(df.head(6)) # as seis primeiras linhas
print()
print(df.head(2)) # as duas primeiras linhas

# Indexação booleana da Data Frame a partir de certos parâmetros
print()
print(df[df['country'] == 'Brazil']) # Seleciona apenas os dados que contenham a coluna 'country' para o parâmetro 'Brazil'
print()
print(df[df['total_litres_of_pure_alcohol'] >= 12]) # Seleciona apenas os dados em que a coluna 'total_litres_of_pure_alcohol' é maior que 12

# Função .sort_values(by='parâmetro', ascending = False) permite o ordenamento da Data Frame a partir de uma coluna especifica em ordens determinadas
# by --> coluna parâmetro / ascending --> True (menor para maior), False(maior para menor)
# É possível usar a função .head() para exibir os cinco primeiros resultados ou .head(numero) para quaisquer resultados
print()
print(df.sort_values(by='wine_servings', ascending = False).head())
```

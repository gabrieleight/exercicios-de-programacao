## Questão 01

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style='darkgrid')
```

```python
carros = sns.load_dataset('mpg')
carros.head()
```

## Questão 02

```python
plt.figure(figsize=(15,5))
plt.title('Relação entre aceleração e consumo de combustível')
sns.scatterplot(x='mpg', y='acceleration', hue='origin', data=carros)
plt.show()
```

#### Existe uma baixa correlação positiva

## Questão 03

```python
plt.figure(figsize=(15,5))
plt.title('Relação entre aceleração e consumo de combustível')
sns.scatterplot(x='mpg', y='acceleration', hue='model_year', data=carros)
plt.show()
```

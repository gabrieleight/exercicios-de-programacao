```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style='darkgrid')
```

## Exemplo 01

```python
quarto_uri = 'https://leadfortaleza.com.br/ead2pcd/conteudo/tmp/myopenolat_1_103418176760922/aula/arquivos/dados.txt'
quarto = pd.read_csv(quarto_uri)
quarto.head()
```

```python
plt.figure(figsize=(15,5))
plt.title('Variação da temperatura ao longo do tempo')
sns.scatterplot(x='date', y='Temperature', data=quarto)
plt.show()
```

```python
quarto['date'] = pd.to_datetime(quarto['date'])
plt.figure(figsize=(15,5))
plt.title('Variação da temperatura de um quarto ao longo do tempo')
sns.lineplot(x='date', y='Temperature', data=quarto)
plt.xlabel('Data')
plt.ylabel('Temperatura (ºC)\n')
plt.show()
```

## Exemplo 02

```python
hiv_uri = 'https://raw.githubusercontent.com/cabrau/curso-analise-visualizacao-dados-python/master/hiv_data.txt'
hiv = pd.read_csv(hiv_uri)
hiv.head()
```

```python
plt.figure(figsize=(15,5))
plt.title('Variação da taxa média de HIV global\n')
sns.lineplot(x='date', y='HIV Rate', data=hiv)
plt.xlabel('Data (anos)\n')
plt.ylabel('Taxa de HIV')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.title('Variação da taxa média de HIV global para níveis de renda\n')
sns.lineplot(x='date', y='HIV Rate', hue='Income Level', ci=None, data=hiv)
plt.xlabel('Data (anos)\n')
plt.ylabel('Taxa de HIV')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.title('Variação da taxa média de HIV global para cada país\n')
sns.lineplot(x='date', y='HIV Rate', hue='country', legend=False, ci=None, data=hiv)
plt.xlabel('Data (anos)\n')
plt.ylabel('Taxa de HIV')
plt.show()
```

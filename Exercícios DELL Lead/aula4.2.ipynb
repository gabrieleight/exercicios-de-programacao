```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style='darkgrid')
```

## Exemplo 01

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
# JointGrid (não permite a utilização do matplotlib)
sns.jointplot(x="sepal_length", y="petal_length", kind = 'reg', data=iris)
plt.show()
```

```python
# JointGrid (não permite a utilização do matplotlib)
sns.jointplot(x="sepal_length", y="petal_length", kind = 'hex', color = 'black', data=iris)
plt.show()
```

```python
# JointGrid (não permite a utilização do matplotlib)
sns.jointplot(x="sepal_length", y="petal_length", kind = 'kde', data=iris)
plt.show()
```

## Exemplo 02

```python
# PairGrid (não permite a utilização do matplotlib)
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris, kind = 'reg')
plt.show()
```

```python
sns.pairplot(iris, hue='species')
plt.show()
```
